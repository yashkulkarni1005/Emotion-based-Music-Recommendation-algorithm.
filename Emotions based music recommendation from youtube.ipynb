```python
#In this Cell based on emotions the songs will be played from the playlist present on personel computer!!

from keras.preprocessing import image
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import os
from pygame import mixer
from keras.models import model_from_json

emotion_label_to_text = {0:'anger', 1:'disgust', 2:'fear', 3:'happy', 4: 'sad', 5: 'surprise', 6: 'neutral'}

json_file = open('files/emotion_model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
emotion_model = model_from_json(loaded_model_json)

emotion_model.load_weights("files/emotion2_model.h5")

img = image.load_img("files/surprise.jpg",target_size=(48,48,3))

         
          
img = cv2.cvtColor(np.float32(img), cv2.COLOR_RGB2GRAY)
          
img = np.array(img, dtype='uint8')
img = np.asarray(img)
plt.imshow(img)
img = np.expand_dims(img, axis=0)
emotion_prediction = emotion_model.predict(img)
#print(np.argmax(emotion_prediction))
emotion_code=np.argmax(emotion_prediction)

text=emotion_label_to_text[emotion_code]
print("Current Emotion is: ",text)
if text == 'anger':
    randomfile = random.choice(os.listdir("files/angry_songs"))
    print('You are angry !!!! please calm down:) ,I will play song for you :' + randomfile)
    file = ('files/angry_songs/' + randomfile)

    
if text == 'disgust':
    randomfile = random.choice(os.listdir("files/disgust_songs"))
    print('You are smiling :) ,I playing special song for you: ' + randomfile)
    file = ('files/disgust_songs/' + randomfile)

if text == 'fear':
    randomfile = random.choice(os.listdir("files/fear_songs"))
    print('You have fear of something ,I playing song for you: ' + randomfile)
    file = ('files/fear_songs/' + randomfile)


if text == 'sad':
    randomfile = random.choice(os.listdir("files/sad_songs"))
    print('You are sad,dont worry:) ,I playing song for you: ' + randomfile)
    file = ('files/sad_songs/' + randomfile)

    
if text == 'happy':
    randomfile = random.choice(os.listdir("files/happy_songs"))
    print('You are happy:) ,enjoy the song : ' + randomfile)
    file = ('files/happy_songs/' + randomfile)

    
if text == 'surprise':
    randomfile = random.choice(os.listdir("files/surprise_songs"))
    print('You look surprised:) , song for you is: ' + randomfile)
    file = ('files/surprise_songs/' + randomfile)

    
if text == 'neutral':
    randomfile = random.choice(os.listdir("files/neutral_songs"))
    print('You are very neutral:) , song for your neutral emotion is : ' + randomfile)
    file = ('files/neutral_songs/' + randomfile)
    

mixer.init()
mixer.music.load(file)
mixer.music.play()
```

```python
##In this Cell based on emotions the songs will be played from the youtube based on the language preference!!
from keras.preprocessing import image
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import os
from pygame import mixer
from keras.models import model_from_json
import pywhatkit

emotion_label_to_text = {0:'anger', 1:'disgust', 2:'fear', 3:'happy', 4: 'sad', 5: 'surprise', 6: 'neutral'}

json_file = open('files/emotion_model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
emotion_model = model_from_json(loaded_model_json)

emotion_model.load_weights("files/emotion2_model.h5")

img = image.load_img("images/img1.jpg",target_size=(48,48,3))

         
          
img = cv2.cvtColor(np.float32(img), cv2.COLOR_RGB2GRAY)
          
img = np.array(img, dtype='uint8')
img = np.asarray(img)
plt.imshow(img)
img = np.expand_dims(img, axis=0)
emotion_prediction = emotion_model.predict(img)
#print(np.argmax(emotion_prediction))
emotion_code=np.argmax(emotion_prediction)

text=emotion_label_to_text[emotion_code]
print("Current Emotion is: ",text)
language=input("Please enter the preferred language : ")
songs=[]
if text == 'anger':
    if language.lower()=='hindi':
        songs=['Sadda_haq','Jee_karda','apna_time_ayega','Yalgaar','Get_ready_to_fight']
    if language.lower()=='english':
        songs=['imagine_dragons','Yummy','rockstar','Believer','Let_me_love_you']
        
elif text == 'happy':
    if language.lower()=='hindi':
        songs=['Gallan_Goodiyan','Ilahi','galiyaan','Raabta','DIl_diya_gallan']
    if language.lower()=='english':
        songs=['wake_me_up','just_the_way_you_are','I_wanna_know','baby_baby','Dynamite']

elif text == 'sad':
    if language.lower()=='hindi':
        songs=['Sun_raha_hai_na_tu','Milne_hai_mujhse_aayi','Samjhawan','shayad','Galliyaan']
    if language.lower()=='english':
        songs=['That_should_be_me','Bruno_mars_Talking_to_the_mooon','shallow','let_me_down_slowly','into_your_arms']
        
elif text == 'neutral':
    if language.lower()=='hindi':
        songs=['dil_kyun_ye_mera','tum_hi_ho_bandhu','Ghungroo','Locha_e_ulfakta','tum_hi_ho_bandhu']
    if language.lower()=='english':
        songs=['shape_of_you','Lean_on','Alone','Faded','Safari','The_Nights']

elif text == 'surprise':
    if language.lower()=='hindi':
        songs=['Chand_si_farish','Mitwa','Kuda_jaane','Dus_bahane','Dil_chata_hai']
    if language.lower()=='english':
        songs=['the_nights','Astronaut_in_the_ocean','FRIENDS','Cheap_thrils','cradles']
        
elif text == 'fear':
    if language.lower()=='hindi':
        songs=['Wafa_ne_bewafai','Lo_mana_liya','Muskurane','Naina','Channa_mereya']
    if language.lower()=='english':
        songs=['Perfect','Shape_of_you','We_dont_talk_anymore','Watermelon_sugar','Darkside']

        
elif text == 'disgust':
    if language.lower()=='hindi':
        songs=['Galliyaan','Lo_mana_liya','Sajdaa','Naina','Channa_mereya']
    if language.lower()=='english':
        songs=['Perfect','Shape_of_you','We_dont_talk_anymore','Watermelon_sugar','Darkside']

else:
    
    print("No Emotion detected.Express your emotions")

song = random.choice(songs)
print("Enjoy the song : "+song)
    
pywhatkit.playonyt(f"https://www.youtube.com/results?search_query=song+{language}+{song}");
```
